Calculate simple moving averages and exponential moving averages for a given stock
price dataset

```python
import pandas as pd

df = pd.read_csv('indexProcessed.csv')
```

```python
# Calculate the Simple Moving Average (SMA)
df['SMA'] = df['Close'].rolling(window=3).mean() # 3-day SMA
df['SMA']
```

```python
# Calculate the Exponential Moving Average (EMA)
df['EMA'] = df['Close'].ewm(span=3, adjust=False).mean() # 3-day EMA
print(df['EMA'])
```
