Perform a Monte Carlo simulation to forecast the future stock prices of a company.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
S0 = 100 # Initial stock price
mu = 0.05 # Expected return
sigma = 0.2 # Volatility
T = 1.0 # Time period (in years)
dt = 1/252 # Time step (daily)
N = int(T/dt) # Number of time steps
num_simulations = 10 # Number of simulations
```

```python
# Simulation
np.random.seed(42) # For reproducibility
prices = np.zeros((N, num_simulations))
prices[0] = S0
```

```python
for t in range(1, N):
    # Generate random normal values for each simulation
    Z = np.random.standard_normal(num_simulations)
    # Calculate the price at each step
    prices[t] = prices[t-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.plot(prices, lw=0.5)
plt.title('Monte Carlo Simulation of Stock Prices')
plt.xlabel('Time Steps')
plt.ylabel('Stock Price')
plt.show()
```

```python
# Calculate the expected future stock price
expected_price = np.mean(prices[-1])
print(f"Expected future stock price after {T} year(s): ${expected_price:.2f}")
```
